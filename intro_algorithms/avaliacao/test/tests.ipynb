```python
#importando os dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\luis\code\python\pos-grad\Introducao_a_algoritmos\avaliacao\data\public_servant_purchases.csv")

df.shape
```

```python
import re
df = df[df['nome_do_funcionario'].apply(lambda x: bool(re.match(r'^[A-Za-z\s]+$', str(x))))]

df
```

```python
df_new = df.dropna(subset=["nome_do_funcionario"])
df_new
```

```python
#detectar compras duplicadas

duplicates_date = df_new['data_da_compra'].duplicated()
duplicates_date.value_counts()

duplicates_true = df_new[duplicates_date].loc[:, ['data_da_compra', 'nome_do_funcionario', 'valor_em_real']]
duplicates_true = duplicates_true.sort_values(by='data_da_compra')
duplicates_true
```

```python
#valores acima de 1000

df_acima_1000=df_new.loc[df["valor_em_real"] > 1000]
df_acima_1000

```

```python
#Identificar compras fora do horário comercial (antes de 08:00 ou depois de 18:00)
df_new['horario_da_compra'] = pd.to_datetime(df['data_da_compra']).dt.time
df_new['dia_da_semana'] = pd.to_datetime(df['data_da_compra']).dt.day_name()
df_new

df_fora_do_horario_comercial = df_new.loc[ (df_new['horario_da_compra'] < pd.to_datetime('08:00').time()) | (df_new['horario_da_compra'] > pd.to_datetime('18:00').time())]
df_fora_do_horario_comercial
```

```python
#Organizar compras por servidor usando dicionarios

compras_por_funcionario = df_new.groupby('nome_do_funcionario')["valor_em_real"].apply(list).to_dict()
compras_por_funcionario
```
